```python
import numpy as np
import matplotlib.pyplot as plt

import librosa
import librosa.display

def plot(path):
    y, sr = librosa.load(path)
    D = librosa.stft(y)  # STFT of y
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    plt.figure()
    librosa.display.specshow(S_db)
    plt.colorbar()
```

```python
plot("speech.wav")
```

```python
plot("output/16_10_2023_08_50_02/A female reporter is speaking full of emotion.wav")
```
